## **Asymmetric**
<hr>

 - Uses a key pair, one for `enc` and `desc`. The keys are termed as `Public Key` and `Private Key`. They `Public Key` is used for the `Encryption` and `Private Key` is used for `Descryption`.

 - `Public Key` is distributed publicly but should never leak the `Private Key`.

 - Every user in the network has their own set of key pairs. 

 - `Asymmetric` relies on mathematical receipe know as `Trapdoor Functions` that uses very large `Prime Numbers`.

 - There is no threat because of sharing the `key` is done with 3-way handshake

The number of Unique Keys required to be generated for `N` users is given by `N*2`. 

### **Trapdoor Functions**
<hr>

 - `Trapdoor Function` is very easy to compute in one direction, yet is very difficult to compute in the opposite direction.

**Whole Idea of Trapdoor Function**

As it uses only `Prime Numbers` for calulation, the multiplication or any arithmetic operation between any large `Prime Numbers` can be done easy but finding the exact factors that resulted the number is very difficult. Therefore, `Trapdoor Function` is best for `Encryption` and `Decryption`. 

For Example, `P1*P2 = N` i.e. `5077*3089 = 15682853`. There might be many factors associated with the `N` but finding the exact ones which resulted `N` is very resourceful and time consuming. 

### **Working of Asymmetric Algo**
<hr>

First, both the users will generate their `Key Pair`. Then there will be the exchange of `Public Keys` between the users. The `Public Key` is transmitted publicly.

Then, the `Sender` `Encrypts` the `Plain Text` message with the `Public Key` of `Receriver`. Then the `Encrypted` messsage i.e. `Cipher Text` is send to the `Receiver` via the internet. 

Now, with the help of `Private Key` of the `Receiver` the message is `Decrypted` into the original `Plain Text` that was sent. 

**Public and Private Key**

The key pair is generated in such a fashion that with the help of the generated `Public Key`, `Private Key` is generated. Anything that is `Encrypted` with the help of `Public Key` can be `Decrypted` with the help of `Private key` that was created using the `Public Key`.

## Hybrid Cryptography
<hr>

In real world, it is very difficul to use only one among `Symmetric` and `Asymmetric` because if we use `Symmetric` we need to be very careful about sharing the `Key` and if we use `Asymmetric` it is computationally slow. 

Therefore, a combination of both is used in real world scenario to use the Advantage of both the types.

**The Working of Hybrid Cryptography**

The sender leverages the `Symmetric` `Encyption` whereas the `Receiver` uses the `Asymmetric` `Decryption`.

Okay,

At first, the `Public Key` of `Receiver` is sent to the `Sender`. Now, the `Sender` will encrypt it's `Symmetric Key` with the `Receiver's Public Key` and then send it to the receiver.  

Once the receiver receives the `Encrypted` shared key of the `Sender` it will be stored safely for future communication. As now, receiver's `Private Key` is used to `Decrypt` the `Encrypted` `Shared Key` to it's original format. Therefore, `Sender` will have all together two keys i.e. `Receiver's Public Key` and `Symmetric Key` and `Receiver` will have three key's i.e `Sender's Symmetric Key`, `Receiver's Private Key` and `Receiver's Public Key`.

Now, for secured transmission of message `Sender` can encrypt the plain text with it's `Shared Key` then the encrypted message is send to the `Receiver`. Once the message is received by the `Receiver`, it can use `Shared key` of `Sender` to decrypt the message.

## **Hashing**
<hr>

Before diving into the main content, we need to know the difference between `Encryption` and `Hashing`. 

### **Difference Between Hashing and Encryption**

**Encryption**

- It is a two way function (Here, both the types are considered two-way function i.e. `Symmetric` and `Asymmetric`). We can use the same key to `Encrypt` and `Decrypt`. 

**Hashing**

- Hashing is a one way function. We can `Encrypt` but `Decryption` is not possible (Irreversible). It must produce unique output for each input. But, the `Hashing` should generate the same output for the same input each time (So that we can compare the hash and find if the data has lost it's `Integrity` or not). Even if there's small change in the `Input` the `Hash` of the input would be completely different.

- Hashing takes variable-length input and always returns fixed-length output.

- Hashing is mainly used for data integrity checks, password storage, and blockchain technology. 

- Hashing Algorithms are very sensitive to small changes in the data. If we compare the `Hash` of the original `Input` with the `Hash` of the changed `Input` the `Hashes` will be different.  

- Hashing is used to ensure `Integrity` of data i.e. Data is not modified in between there's sharing of data.
<hr>

### **How does hashing work?**

At, first the `Input` goes through the `Hashing` algorithm which is a one way function. After passing through the `One Way Function` it gives output which is known as `Message Digest` is called `Hash` of the given input. This way a `Hash` is generated for a given input. 

`Hashing` is used to store sensitive infos for example, `password`. `Hash` of the given password is saved into the database for that particular user. When the user tries to sign in we will first create `Hash` for the given password and then compare the `Hashes` for that user. If matched we can authorize. 

Along with `Hashing` algorithm a extra `Salt` along with the `Hashing` Algorithm. The `Salt` value is stored along with the `Hash` value for each user in the database. 

**Common One Way Function Example**

- Finding the exact prime factors of a number

-  Discrete Logarithm

### **Asymmetric Cryptography with Hashing**

First exchange of `Public Keys`. Sender encrypts the text with the `Public Key` of `Receiver`, and for the same plain text a `Hash` is generated using a `Hashing` algorithm. Then `Encrypt` the generated `Hash` value with the `Private key` of the `Sender`. The resut is the `Digigtal Signature` for this message (plain text). 

Then the both `Encrypted` message and `Encrypted` `Hash` arrives to the `Receiver`. Firstly, `Receiver's` `Private Key` is used to `Decrypt` the `Encrypted` plain text. Then, the plain text is run through the same `Hashing` algorithm to generate the `Hash`. Then, again the `Public Key` of `Sender` to `Decrypt` the `Hash`. Once, both the `Hash` are ready, they are compared to see if they still hold the `Integrity`. If the `Hash` value are same then it verifies the `Digital Signature` else not.

## **Common Symmetric Algorithms**
<hr>

### **Block Based**

**DES (Data Encryption Standard)** 

In 1975, introduces as the first block based algorithm. It `Encrypt`s 64-bit block of data using a 56-bit key. 

Of those 65 bits only 56 bits are used to stored information. Rest of the 8-bit contains the `parity` informations. 

`Parity` bit are check bit which ensures that the 56-bit are correct. 

It goes through 16 rounds of processing. It was cracked in 1988 in 3 days. 

<img src='./Notes_Images/c23.png'>

**3DES**

Introduced in 1978 as quick fix for `DES`. 64-bit block of data using a 112-bit or 168-bit key; always three iterations done with key. 

Goes through 48 rounds of `Encryption` processing. 

<img src='./Notes_Images/c24.png'>

**AES (Advanced Encryption Standard)** 

Introduces in the 2000 to replace `DES`; originally called `Rijndael`. It is the new current standard till date. 

Uses 128-bit block of data, using 128-bit, 192-bit, or 256-bit key. Fewer rounds of `Enc` due to longer length of the key. 

<img src='./Notes_Images/c25.png'>

**RC5 and RC6 (Block Based)**

To overcome the weakness of `RC4`; built faster based on block cipher. 

Uses key size of (0 to 2040 bits); variable block size (32, 64, 128 bits). 

Few rounds of `Enc` (0 to 255). 

They are stable, used till date; not reported any vulnerable to attacks. 

<img src='./Notes_Images/c27.png'>

## **Fundamentals of Public Key Infrastructure (PKI)**
<hr>

It is an Universal infrastructure to establish trust within the communicating `Entities` or `Devices`.

Uses `Asymmetric` Encryption for secure communication. 

To maintain the security for the `3 States of Data` we implement the `PKI` framework. 

<img src='./Notes_Images/c39.png'>


### **Three States of Data**

**Data at Rest** : Stored on some type of storage device or system. 

**Data at Transit** : Moving across the network. End to End `Encryption`. 

**Data at Use** :  Consumed or active. 

<hr>

At the heart of `PKI` there are two main concepts:

### **Digital Signature**

The `Private Key` of the `Sender` is used to create a `Digital Signature` i.e. Hash of the Original Plain Text and corresponding `Public Key` of the `Sender` is used to verify the signature. 

The `Digital Signature` on a message provides a unique electronic biniding of the identify of the signer to the origin of the message. It provides the proof of the mesage origin and method to verify the integrity of the message.

### **Digital Certificate**

Similary, `Digital Certficates` is an additional way of documents proving identity. The difference between `Digital Signature` and `Digital Certificate` is that the `Digital Certificate` binds the `Digital Signature` to the object while the `Digital Signature` must ensure that the data and the information remains secure from the moment it's sent. 

### **PKI (Publick Key Infrastructure) Components**

**Certificate Management System** : It allows us to manage all aspects of the lifecyle of `Certificates` which includes `Validation`, `Deployment`, `Revocation` and `Renewal` to provide uniterputed service. 

**End User** : Can be a person or entity requesting and using secure digital communications. 

**Certificate Authority (CA)** : It operates as a trusted introducer. It is primarily responsible for `Issuing`, `Revoking` and `Distributing Digital Certificates`. A `CA` is voucing (Confirms) for the certificate's owner's identity in the same way as having a government issued `ID` such as `Passport`. There are many trusted third part organizations who acts as `CA`s i.e. `GlobalSign`, `Verisign`, `Entrust`, `GoDaddy`, `AWS`, `IdenTrust`, `DigiCert`. 

If we wish to create our own `Certificate Authority` we can.

### **Understanding PKI with an Example**

Suppose, Brian and Tasha are about to communicate with each other using `Asymmetric`. Firstly they exchange the `Public Keys` then `Brian` encrypts the message and sends it to `Tash`. `Tash` with her `Private Key` Decrypts the message. 

Here, even if the hacker gets access to the shared `Public Key` via unsecured channels won't be able to create a damage becuase of `Private Key`. But, what if `Someone` was pretending to be `Tasha` and send `Public Key` that don't even belong to `Tasha`?? `Brian` only thought he was communicating to `Tash` the whole time. 

This is where the whole idea of `Digital Certificate` comes into the play. 

`Digital Certificate` bind an identity to a `Public Key`. It basically verifys the `Sender` or the `Receiver` using a `Cerfificate Authority (CA)`

**This Time Communicating with Confidence (Digital Certificates)**

To create a `Digital Certificate` `Tash` will create new message called `Certified Signing Request` or `CSR` which contains her `Public Key`, `Name`, `Email Address` and `Other Identifying Infos`. 

This request gets forwarded to the `Registration Authority` or `RA`. The `RA` will take some of the adminstrative burden from the `CA` by verifying `Tasha's` infos. If the `RA` grants the request, `Tasha's` request is now passed to the `CA` to generate the `Digital Certificate`. If the provided infos are authenticated then `CA` will use its `Private Key` to place the `CA's` `Digital Signature` (Hash) on `Tasha's` `Digital Certificate`. 

Then, finally send it directly to `Tasha` to complete the process. The format she receives it in is known as `X509`, which is the international standard for `Digital Certificates`. 

Now, `Brian` can either have `Tasha` send her certificate over, or get it directly from the `CA`. 

Okay, now when the `Encrypted` message is received by `Brian`, he trusts the `Certificate Authority` used by `Tasha`. Now, as `Brian` has the `Digital Certificate` that contains `Tasha's` specifics such as her `Public Keys`, `CA` specifics, and other `Cryptographic Data` that are required to identify `Tasha`. `Brian` uses the `CA's` `Public Key` to `Decrypt` and validate the certificate. Once the `Digital Certificate` is `Decypted`, 

`Brian` can now access the content inside it. `Brian` can now access `Tasha's` `Public Key` without tranfering over the unsecured channel. Then using the `Public Key` `Brian` decrypts the other messages. 

During this whole process, `Brain` got access to `Tasha's` `Public Key` and this time he had a way to validate that the `Public Key` came from `Tasha` by validating her `Digital Certificate`.
